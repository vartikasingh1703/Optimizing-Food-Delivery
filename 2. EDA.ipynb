# Exploratory Data Analysis (EDA)

## Importing the dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv") 
```

## Impact of Weather & Traffic on Delivery Times

```python
#Boxplot: Delivery Time vs. Road Traffic Density

plt.figure(figsize=(10, 6))
traffic_order = ["Low","Medium","High", "Jam"]
sns.boxplot(x='Road_traffic_density', y='Time_taken(min)',order=traffic_order, data=df)
plt.title("Delivery Time vs. Road Traffic Density")
plt.xlabel("Road Traffic Density")
plt.ylabel("Delivery Time (min)")
plt.show()
```

### Insight: Does delivery time increase significantly with higher traffic density?
### Yes 

## Weather Conditions Impacting Late Deliveries

```python
plt.figure(figsize=(10, 6))
condition_order= ["Sunny" , "Cloudy","Windy","Fog","Stormy", "Sandstorms","NaN"]
sns.boxplot(x='Weatherconditions', y='Time_taken(min)',order=condition_order, data=df)
plt.title("Weather Conditions Impact on Delivery Time")
plt.xlabel("Weather Conditions")
plt.ylabel("Delivery Time (min)")
plt.show()
```

### Insight: Do bad weather conditions cause significant delivery delays?
### Yes , but slightly

## Vehicle Type vs. Delivery Speed

```python
#Boxplot: Delivery Time by Vehicle Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Type_of_vehicle', y='Time_taken(min)', data=df)
plt.title("Delivery Time by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Time Taken (min)")
plt.show()
```

 ### Insight: Which vehicles have the fastest delivery times?
 ### Scooter and Electric Scooter

## Impact of Festival Days on Delays

```python
# Histogram: Delivery Time on Festival vs. Non-Festival Days
sns.boxplot(x='Festival', y='Time_taken(min)', data=df, palette="Set2")
plt.title("Delivery Time on Festival vs. Non-Festival Days")
plt.xticks(ticks=[0, 1], labels=["Festival", "Non-Festival"])
plt.xlabel("Festival Day")
plt.ylabel("Delivery Time (minutes)")
plt.show()


```

### Insight: Do festival days significantly impact delivery time?
### Yes

## Are Older Delivery Agents Faster or Slower?

```python
# Scatterplot: Age vs. Delivery Time
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Delivery_person_Age", y="Time_taken(min)", data=df, color="purple")
plt.title("Delivery Person Age vs. Delivery Time")
plt.xlabel("Delivery Person Age")
plt.ylabel("Delivery Time (minutes)")
plt.show()

```

### Insight: Is there a trend in speed based on age?
### Slightly


## Effect of Multiple Deliveries on Time Taken

```python
#Boxplot: Multiple Deliveries vs. Time Taken
plt.figure(figsize=(10, 6))
no_of_deliveries = [0 ,1 ,2 ,3 ,'NaN']
sns.boxplot(x='multiple_deliveries', y='Time_taken(min)',order=no_of_deliveries, data=df)
plt.title("Effect of Multiple Deliveries on Time Taken")
plt.xlabel("Number of Multiple Deliveries")
plt.ylabel("Delivery Time (min)")
plt.show()
```

 ### Insight: Do drivers taking multiple deliveries take significantly longer?
 ### Yes

## Correlation Heatmap Between Features


```python
import numpy as np

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

### Insight: Which features are most strongly correlated with delivery time?
### 1.Road_traffic_density
### 2.multiple deliveries
### 3.Distance_km
### 4.Festival
### 5.Delivery_person_Age
